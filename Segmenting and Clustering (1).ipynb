```python
# Create dataframe by using pandas
df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df[0]
```

```python
df=df[0]
```

```python
df.head()
```

```python
# drop cells with a borough that is Not assigned
index_NA=df[(df['Borough']=='Not assigned')].index
```

```python
df.drop(index_NA, inplace=True)
```

```python
df.head()
```

```python
#code: If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough
df.loc[df.Neighborhood=='Not assigned','Neighborhood']=df.Neighborhood
```
